# :basketball: 5 - Train combined dataset

## Imports

```python
import shutil
import pandas as pd

from pathlib import Path
from ultralytics import YOLO
```

```python
import torch
print(torch.cuda.is_available())
```

```python
import warnings
warnings.filterwarnings("ignore")
```

## Predefined variables

```python
YAML_PATH = "../data/basketball.yolov8-combined/data.yaml"
RESULTS_DIR = "../results/"

PROJECT_ROOT = Path.cwd().parent
RUNS_DIR = PROJECT_ROOT / "runs" / "combined"

Path(RUNS_DIR).mkdir(parents=True, exist_ok=True)

EPOCHS = 100
IMG_SIZE = 640
BATCH_SIZE = 16
```

## Train combined dataset

```python
# Load YOLOv8n model
yolov8 = YOLO("yolov8n.pt", "detect").to('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Train the model
yolov8.train(
    data=YAML_PATH,
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    project=str(RUNS_DIR),
    name="yolov8n-combined-train",
    augment=False,
    exist_ok=False,
    workers=0 # Set workers to 0 for compatibility and reduce overhead
)
```

## Evaluate base model

```python
# Evaluate the model
metrics = yolov8.val(
    data=YAML_PATH,
    project=str(RUNS_DIR),
    name="yolov8n-combined-valid",
    split="test",
    exist_ok=False
)
```

```python
print("Evaluation Metrics:")
print(metrics)
```

## Save metrics

```python
# Export results to a DataFrame for better visualization
results = metrics.results_dict

table = {
    "mAP50": results['metrics/mAP50(B)'],
    "mAP50-95": results['metrics/mAP50-95(B)'],
    "Precision": results['metrics/precision(B)'],
    "Recall": results['metrics/recall(B)']
}

metrics_df = pd.DataFrame([table])
metrics_df
```

```python
# Save the metrics to a CSV file
metrics_df.to_csv(RESULTS_DIR + "combined_metrics.csv", index=False)
```
